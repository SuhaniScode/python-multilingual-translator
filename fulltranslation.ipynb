```python
from googletrans import Translator
import pandas as pd
# Create a translator object
translator = Translator()
def translate_text(text, src_lang, dest_lang):
    """
    Very simple translation function.
    - text: the text to translate
    - src_lang: source language code (hi, en, or, auto, etc.)
    - dest_lang: target language code (hi, en, or, etc.)
    """
    if not text.strip():
        return ""
    try:
        result = translator.translate(text, src=src_lang, dest=dest_lang)
        return result.text
    except Exception as e:
        print("Error during translation:", e)
        return text  # return original if translation fails
```

```python
def interactive_translate():
    print("Welcome to Simple Translator!")
    print("Type 'quit' anytime to stop.")
    print("Use language codes like: hi = Hindi, en = English, or = Odia, auto = auto-detect\n")
    while True:
        src = input("Source language code (hi/en/or/auto): ").strip()
        if src.lower() == "quit":
            break
        dest = input("Target language code (hi/en/or): ").strip()
        if dest.lower() == "quit":
            break
        text = input("Enter text to translate: ").strip()
        if text.lower() == "quit":
            break
        translated = translate_text(text, src, dest)
        print("Translated:", translated)
        print("-" * 40)
```

```python
def translate_csv_column(input_csv, column_name, src_lang, dest_lang):
    """
    Reads a CSV file, translates one column, saves a new CSV.
    """
    df = pd.read_csv(input_csv)

    if column_name not in df.columns:
        print("Column not found!")
        return
    new_col = column_name + "_translated"
    translations = []
    for text in df[column_name]:
        translated = translate_text(str(text), src_lang, dest_lang)
        translations.append(translated)
    df[new_col] = translations
    output_name = input_csv.replace(".csv", "_translated.csv")
    df.to_csv(output_name, index=False)
    print("Translation completed. File saved as:", output_name)
```

```python
print(translate_text("नमस्ते", "hi", "en"))
print(translate_text("Hello, how are you?", "en", "hi"))
print(translate_text("ମୋର ନାମ ସୁହାନୀ", "or", "en"))
```
